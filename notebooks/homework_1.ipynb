```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

# =============================================
# Loading the data
# =============================================

# Read train and test csv files
df_raw_train = pd.read_csv(abspath_curr + 'data/train.csv', header=0)
df_train = df_raw_train.copy(deep=True)

df_raw_test = pd.read_csv(abspath_curr + 'data/test.csv', header=0)
df_test = df_raw_test.copy(deep=True)

# The target column for this competition is trip_duration (in seconds)
target = 'trip_duration'

print("df_train:", df_train.shape)
print("df_test:", df_test.shape)
display(df_train.head())

# =============================================
# Splitting the data
# =============================================

# Split into 80% train and 20% validation (same as the case study)
df_train, df_val = train_test_split(df_train, train_size=0.8, random_state=random_seed)
df_train, df_val = df_train.reset_index(drop=True), df_val.reset_index(drop=True)

# =============================================
# Handling uncommon features
# =============================================

# Check which columns are shared across train, val, and test
df_common_var = common_var_checker(df_train, df_val, df_test, target)

# Find columns that only exist in one dataset
# (e.g., dropoff_datetime only appears in train data, not in test)
uncommon_feature_train_not_val_test = np.setdiff1d(df_train.columns, df_common_var['common var'])
uncommon_feature_val_not_train_test = np.setdiff1d(df_val.columns, df_common_var['common var'])
uncommon_feature_test_not_train_val = np.setdiff1d(df_test.columns, df_common_var['common var'])

print("Uncommon in train:", uncommon_feature_train_not_val_test)

# Drop those uncommon columns so all three datasets have same features
df_train = df_train.drop(columns=uncommon_feature_train_not_val_test)
df_val = df_val.drop(columns=uncommon_feature_val_not_train_test)
df_test = df_test.drop(columns=uncommon_feature_test_not_train_val)

# =============================================
# Handling identifiers
# =============================================

# Combine all data to find identifier columns
df = pd.concat([df_train, df_val, df_test], sort=False)

# id column has unique value for every row, so it is an identifier
df_id = id_checker(df)
print("Identifiers:", list(df_id.columns))

# Remove identifier columns since they don't help prediction
df_train.drop(columns=np.intersect1d(df_id.columns, df_train.columns), inplace=True)
df_val.drop(columns=np.intersect1d(df_id.columns, df_val.columns), inplace=True)
df_test.drop(columns=np.intersect1d(df_id.columns, df_test.columns), inplace=True)

# =============================================
# Handling date time variables
# =============================================

# pickup_datetime needs to be converted into numeric features
datetime_vars = ['pickup_datetime']

# This splits pickup_datetime into year, month, day, hour, minute, second
df_train = datetime_transformer(df_train, datetime_vars)
df_val = datetime_transformer(df_val, datetime_vars)
df_test = datetime_transformer(df_test, datetime_vars)

# =============================================
# Feature engineering
# =============================================

# I added some extra features to try to beat the baseline score
def add_features(df):
    """Create new features from the existing coordinate and datetime columns."""
    d = df.copy(deep=True)

    # Calculate the actual distance between pickup and dropoff using haversine formula
    # Reference: https://en.wikipedia.org/wiki/Haversine_formula
    R = 6371  # Earth radius in km
    lat1 = np.radians(d['pickup_latitude'])
    lat2 = np.radians(d['dropoff_latitude'])
    dlat = np.radians(d['dropoff_latitude'] - d['pickup_latitude'])
    dlon = np.radians(d['dropoff_longitude'] - d['pickup_longitude'])
    a = np.sin(dlat / 2) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2) ** 2
    d['haversine_distance'] = 2 * R * np.arcsin(np.sqrt(a))

    # Simple coordinate differences (kind of like Manhattan distance idea)
    d['abs_diff_longitude'] = np.abs(d['dropoff_longitude'] - d['pickup_longitude'])
    d['abs_diff_latitude'] = np.abs(d['dropoff_latitude'] - d['pickup_latitude'])

    # Day of week might matter because weekday vs weekend traffic is different
    d['pickup_datetime_dayofweek'] = pd.to_datetime(
        d[['pickup_datetime_year', 'pickup_datetime_month', 'pickup_datetime_day']].rename(
            columns={'pickup_datetime_year': 'year', 'pickup_datetime_month': 'month', 'pickup_datetime_day': 'day'}
        )
    ).dt.dayofweek

    return d

df_train = add_features(df_train)
df_val = add_features(df_val)
df_test = add_features(df_test)

print("Columns after feature engineering:", list(df_train.columns))

# =============================================
# Handling missing data
# =============================================

df = pd.concat([df_train, df_val, df_test], sort=False)

df_nan = nan_checker(df)
print("Missing values:")
display(df_nan)

# Only impute float columns that have missing values
df_miss = df_nan[df_nan['dtype'] == 'float64'].reset_index(drop=True)

# Separate back into train, val, test
df_train = df.iloc[:df_train.shape[0], :]
df_val = df.iloc[df_train.shape[0]:df_train.shape[0] + df_val.shape[0], :]
df_test = df.iloc[df_train.shape[0] + df_val.shape[0]:, :]

# Use mean imputation (same approach as case study)
if len(df_miss['var']) > 0:
    si = SimpleImputer(missing_values=np.nan, strategy='mean')
    df_train[df_miss['var']] = si.fit_transform(df_train[df_miss['var']])
    df_val[df_miss['var']] = si.transform(df_val[df_miss['var']])
    df_test[df_miss['var']] = si.transform(df_test[df_miss['var']])

# =============================================
# Encoding the data
# =============================================

df = pd.concat([df_train, df_val, df_test], sort=False)

# Check for categorical columns (store_and_fwd_flag is the only one here)
df_cat = cat_var_checker(df)
print("Categorical variables:")
display(df_cat)

# One-hot encode categorical features
if len(df_cat) > 0:
    df = pd.get_dummies(df, columns=np.setdiff1d(df_cat['var'], [target]))

df_train = df.iloc[:df_train.shape[0], :]
df_val = df.iloc[df_train.shape[0]:df_train.shape[0] + df_val.shape[0], :]
df_test = df.iloc[df_train.shape[0] + df_val.shape[0]:, :]

print("df_train shape after encoding:", df_train.shape)

# =============================================
# Splitting the feature and target
# =============================================

X_train = df_train[np.setdiff1d(df_train.columns, [target])].values
X_val = df_val[np.setdiff1d(df_val.columns, [target])].values
X_test = df_test[np.setdiff1d(df_test.columns, [target])].values

y_train = df_train[target].values
y_val = df_val[target].values
y_test = df_test[target].values

# =============================================
# Scaling the data
# =============================================

ss = StandardScaler()

# Standardize features first
X_train = ss.fit_transform(X_train)
X_val = ss.transform(X_val)
X_test = ss.transform(X_test)

# Then standardize target (refit ss on target so inverse_transform works later)
y_train = ss.fit_transform(y_train.reshape(-1, 1)).reshape(-1)
y_val = ss.transform(y_val.reshape(-1, 1)).reshape(-1)
y_test = ss.transform(y_test.reshape(-1, 1)).reshape(-1)
```

# Hyperparameter Tuning

```python
from sklearn.linear_model import SGDRegressor, Ridge, Lasso
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV

# =============================================
# Creating the dictionary of the models
# =============================================

# I chose Ridge and Lasso in addition to SGD to compare different regularization approaches
models = {
    'sgd': SGDRegressor(random_state=random_seed),
    'ridge': Ridge(),
    'lasso': Lasso(random_state=random_seed)
}

# =============================================
# Creating the dictionary of the pipelines
# =============================================

pipes = {}
for acronym, model in models.items():
    pipes[acronym] = Pipeline([('model', model)])

# =============================================
# Getting the predefined split cross-validator
# =============================================

X_train_val, y_train_val, ps = get_train_val_ps(X_train, y_train, X_val, y_val)

# =============================================
# Creating the dictionary of the parameter grids
# =============================================

param_grids = {}

# SGD: tuning learning rate and regularization strength
param_grids['sgd'] = [{'model__eta0': [0.0001, 0.001, 0.01],
                        'model__alpha': [0.0001, 0.001, 0.01, 0.1]}]

# Ridge: tuning regularization strength (L2 penalty)
param_grids['ridge'] = [{'model__alpha': [0.01, 0.1, 1.0, 10.0]}]

# Lasso: tuning regularization strength (L1 penalty)
param_grids['lasso'] = [{'model__alpha': [0.0001, 0.001, 0.01, 0.1]}]

# =============================================
# Creating the directory for the cv results
# =============================================

directory = os.path.dirname(abspath_curr + 'result/cv_results/GridSearchCV/')
if not os.path.exists(directory):
    os.makedirs(directory)

# =============================================
# Tuning the hyperparameters
# =============================================

best_score_params_estimator_gs = []

for acronym in pipes.keys():
    print(f"Training {acronym}...")

    # Run grid search with neg MSE scoring (same settings as case study)
    gs = GridSearchCV(estimator=pipes[acronym],
                      param_grid=param_grids[acronym],
                      scoring='neg_mean_squared_error',
                      n_jobs=2,
                      cv=ps,
                      return_train_score=True)

    gs = gs.fit(X_train_val, y_train_val)

    # Save the best result for this model
    best_score_params_estimator_gs.append([gs.best_score_, gs.best_params_, gs.best_estimator_])

    # Sort and save cv results to csv
    cv_results = pd.DataFrame.from_dict(gs.cv_results_).sort_values(by=['rank_test_score', 'std_test_score'])

    important_columns = ['rank_test_score',
                         'mean_test_score',
                         'std_test_score',
                         'mean_train_score',
                         'std_train_score',
                         'mean_fit_time',
                         'std_fit_time',
                         'mean_score_time',
                         'std_score_time']

    cv_results = cv_results[important_columns + sorted(list(set(cv_results.columns) - set(important_columns)))]
    cv_results.to_csv(path_or_buf=abspath_curr + 'result/cv_results/GridSearchCV/' + acronym + '.csv', index=False)

# Sort by best score (highest = best since neg MSE is closer to 0)
best_score_params_estimator_gs = sorted(best_score_params_estimator_gs, key=lambda x: x[0], reverse=True)

# Show results for all models
pd.DataFrame(best_score_params_estimator_gs, columns=['best_score', 'best_param', 'best_estimator'])
```
